```python
import subprocess

# Get your workspace URL and token
WORKSPACE_URL = "https://dbc-fc5d7400-6f13.cloud.databricks.com"   # ← your workspace URL
TOKEN         = dbutils.notebook.entry_point.getDbutils().notebook()\
                    .getContext().apiToken().get()                   # ← auto-fetches your token

# Configure CLI
result = subprocess.run(
    ["databricks", "configure", "--token"],
    input  = f"{WORKSPACE_URL}\n{TOKEN}\n",
    capture_output = True,
    text   = True
)
print("✅ CLI configured")
```

```python
import subprocess

# Create scope
subprocess.run(
    ["databricks", "secrets", "create-scope",
     "--scope", "app_secrets",
     "--initial-manage-principal", "users"],
    capture_output=True, text=True
)
print("✅ Scope created: app_secrets")

# Store Claude API key
result = subprocess.run(
    ["databricks", "secrets", "put",
     "--scope", "app_secrets",
     "--key",   "claude_api_key",
     "--string-value", "sk-ant-api03-.................."],   # ← your real key here
    capture_output=True, text=True
)
print(f"✅ Secret stored")
print(result.stderr if result.stderr else "No errors")
```
